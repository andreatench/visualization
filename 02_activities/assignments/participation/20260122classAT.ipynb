```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x = np.arange(50) #from 0 to 49
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) #size
ax.plot(x,y1)
ax.plot(x,y2) #adding another line plot
fig.show() #show the entire - if just in python need it
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
#removing tick marks and labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#limiting tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
#tick mark intervals
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
#rotating axis labels 

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
```

```python
#modifying axis title fonts


font1 = {'family':'serif','color':'indigo'}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')

plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```

```python
font1 = {'family':'serif','color':'indigo'}
font2 = {'family':'serif','color':'hotpink'}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')

plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
plt.ylabel('Shiny New Y Axis!', fontsize = 18, fontdict = font2)
```
